# Transforming the dataset

```python
num_vars = ['temp', 'atemp', 'hum', 'windspeed','cnt']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df_train.columns
```

```python
df_train
```

```python
X_train=df_train.drop(columns=['cnt','holiday','hum','weekday_1','workingday','weekday_2','weekday_3','temp','weekday_4','weekday_5','weekday_6','mnth_2','mnth_3','mnth_4','mnth_5','mnth_6','mnth_7','mnth_8','weathersit_2','mnth_11','mnth_12'])
y_train=df_train['cnt']
```

```python
lm=LinearRegression()
lm.fit(X_train,y_train)
rfe = RFE(lm, n_features_to_select=15)             # running RFE
rfe = rfe.fit(X_train, y_train)
col = X_train.columns[rfe.support_]
col
X_train_rfe = X_train[col]
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
ols_model= sm.OLS(y_train,X_train_rfe).fit() 
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
ols_model.summary()
```

# Checking the Multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

# transforming the testing data

```python
X_test=df_test.drop(columns=['cnt','holiday','hum','weekday_1','workingday','weekday_2','weekday_3','temp','weekday_4','weekday_5','weekday_6','mnth_2','mnth_3','mnth_4','mnth_5','mnth_6','mnth_7','mnth_8','weathersit_2','mnth_11','mnth_12'])
y_test=df_test['cnt']
```
